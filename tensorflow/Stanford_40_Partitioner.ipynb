```python
IM_DIR='/media/red/capstone/data/stanford_40/JPEGImages_partitioned/'

import os
from os import listdir
from os.path import isfile, join
import pdb
import shutil
onlyfiles = [f for f in listdir(IM_DIR) if isfile(join(IM_DIR, f))]

for i, fname in enumerate(onlyfiles):
    print(i)
    name, ext = fname.split('.')
    name = name.rsplit('_', 1)[0]
    print(name)
    os.makedirs(os.path.join(IM_DIR, name), exist_ok=True)
    src = os.path.join(IM_DIR, fname)
    dst = os.path.join(IM_DIR, name, fname)
    shutil.move(src, dst)
```
